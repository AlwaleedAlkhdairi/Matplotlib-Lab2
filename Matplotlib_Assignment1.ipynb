```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
# Create data for the pie chart using a list
gender = ['Male', 'Femal']
values = [25, 40]

# Create a bar chart using plt.pie()
plt.figure(figsize=(8, 5))
plt.pie(values, labels= gender,autopct='%1.2f%%', colors=['#66b3ff','#ff9999'],shadow=True, radius=1, explode = [0.1,0])
plt.xlabel('gender')
plt.ylabel('Values')
plt.title('gender pie Chart using List')
plt.show()

# Create data for the scatter plot using a DataFrame
data = {'X': [10, 20, 30, 40, 50],
        'Y': [25, 45, 15, 30, 50]}
df = pd.DataFrame(data)

# Create a scatter plot using df.plot.scatter()
plt.figure(figsize=(8, 5))
plt.scatter(df['X'], df['Y'], color='red')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter Plot using DataFrame')
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
# Create data for the pie chart using a list
gender = ['Male', 'Femal']
values = [25, 40]

# Create a bar chart using plt.pie()
plt.figure(figsize=(8, 5))
plt.pie(values, labels= gender,autopct='%1.2f%%', colors=['#66b3ff','#ff9999'],shadow=True, radius=1, explode = [0.2,0.2])
plt.xlabel('gender')
plt.ylabel('Values')
plt.title('gender pie Chart using List')
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
data = {'X': [10, 20, 30, 40, 50],
        'Y': [25, 45, 15, 30, 50]}
df = pd.DataFrame(data)

# Create a scatter plot using df.plot.scatter()
plt.figure(figsize=(8, 5))
plt.scatter(df['X'], df['Y'], color='red', marker = 'o', s = 50,alpha =0.4)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter Plot using DataFrame')
plt.show()
```

## Q4: Add the following to your chart:

```python

data = {'X': [10, 20, 30, 40, 50],
        'Y': [25, 45, 15, 30, 50]}
df = pd.DataFrame(data)

# Create a scatter plot using df.plot.scatter()
plt.figure(figsize=(8, 5))
plt.scatter(df['X'], df['Y'], color='red', marker = 'o', s = 50,alpha =0.4)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter Plot using DataFrame', fontsize=16, color='red', loc='left')
plt.show()
```
