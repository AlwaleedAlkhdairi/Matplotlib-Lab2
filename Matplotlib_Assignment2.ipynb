# Q2: Create a one line chart using two arrays. The first array represents x and the second array represents y
| x     |     y    |
| ----- | -------- |
| 1    | 1       |
| 2    | 2       |
| 3    | 3       |
| 4    | 5       |
| 5    | 8       |
| 6    | 13      |
| 7    | 20      |

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 2, 3, 5, 8, 13, 20]

plt.plot(x, y, color='blue', linestyle='-', linewidth=3, markersize=10)

# Add labels and title
plt.xlabel('first array', fontsize=14)
plt.ylabel('second array', fontsize=14)
plt.title('Line Chart', fontsize=16)

# Show the chart
plt.show()
```

# Q3: Create a two lines chart using the same previous arrays. 
But this time:
- Mark each point with a circle.
- Use dashed lines.
- Set the size of the marker (circle) to size 10.
- Set the width of the line to 3.
- Add legends. 'Normal' for array x and 'Fast' for array y.

```python
plt.plot(x, y, marker='o', color='blue', linestyle='--', linewidth=3, markersize=10, label='Fast')
plt.plot(x, x, marker='o', color='red', linestyle='--', linewidth=3, markersize=10, label='Normal')

plt.xlabel('X values', fontsize=14)
plt.ylabel('Y values', fontsize=14)
plt.title('Two Lines Chart', fontsize=16)
plt.legend()

plt.show()
```

# Q4: Display multiple line charts in one row using the charts given below.
- Give each chart a title 
- Add a super title 'my data visulaization assignment'. 

```python
#super title = my data visulaization assignment

#title = First Chart
x_plot1 = np.array([1, 2, 3, 4, 5, 6, 7])
y_plot1 = np.array([1, 1, 2, 3, 5, 8, 13])

#title = Second Chart
x_plot2 = np.array([0, 1, 2, 3, 4, 5, 6])
y_plot2 = np.array([2, 4, 6, 8, 10, 12, 14])

#title = Third Chart
x_plot3 = np.array([0, 1, 3, 4])
y_plot3 = np.array([4, 6, 3, 4])

# Create a figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(10, 5))

axes[0].plot(x_plot1, y_plot1, marker='o', color = 'r')
axes[0].set_title('First Chart')
axes[0].set_xlabel('X values')
axes[0].set_ylabel('Y values')

axes[1].plot(x_plot2, y_plot2, marker='*')
axes[1].set_title('Second Chart')
axes[1].set_xlabel('X values')
axes[1].set_ylabel('Y values')

axes[2].plot(x_plot3, y_plot3, marker='+', color = 'g')
axes[2].set_title('Third Chart')
axes[2].set_xlabel('X values')
axes[2].set_ylabel('Y values')

fig.suptitle('my data visulaization assignment', fontsize=16)

# For better apearence
plt.tight_layout()

# Show the plots
plt.show()
```

# Q5: Define an array that has a values of your top three TV series and define another array to hold the rating of the three TV series out of 10. Create a bar chart using the two arrays
- Set a new color for the bars.
- Change the width of the bars to width 0.5.

```python
# Top my three TV series and their ratings
series = np.array(['peaky blinders', 'Sherlock Holmes', 'The 100'])
ratings = np.array([9.2, 9.5, 8.7])

# Create a bar chart
plt.bar(series, ratings, color=['darkorchid','purple','darkorchid'], width=0.5)

# Add title and labels
plt.title('Top TV Series Ratings (by me)')
plt.xlabel('TV Series')
plt.ylabel('Rating out of 10')

# Show the plot
plt.show()
```

# Q6: Create a scatter plot  using two arrays. The first array represents carAge and the second array represents carSpeed, set a specific color for each dot by using an array of different colors and add labels for the plot
| carAge | carSpeed |
| ----- | -------- |
| 5    | 99       |
| 7    | 86       |
| 8    |87      |
| 7    | 88       |
| 2    | 111       |
| 17    | 86       |
| 2    | 103      |
| 9    | 87       |
| 4    | 94       |
| 11    | 78      |

- Set a new color for each dot by using an array of different colors.
- Change the size of the dots to size 350.
- Add labels for x-axis and y-axis.
- Adjust the transparency of the dots with the alpha argument. let the alpha value be 0.4

```python
car_age = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11])
car_speed = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78])

colors = ['red', 'green', 'blue', 'orange', 'purple', 'yellow', 'cyan', 'magenta', 'pink', 'brown']

plt.scatter(car_age, car_speed, s=350, c=colors, alpha=0.4)
plt.xlabel('Car Age')
plt.ylabel('Car Speed')
plt.title('Car Age vs Car Speed');
```

# Q7: Create a pie chart using two arrays. The first array represents python_libraries and the second array represents your skills out of 10 at each library
| python_libraries |   
| ------------ | 
| NumPy        | 
| Pandas       | 
| Matplotlib   | 
| Seaborn      | 
| Plotly       | 
| Scikit-learn | 

- Set a new color for each wedge by using an array of different colors.
- Pull your favorite library wedge 0.3 from the center of the pie.
- Add a shadow to the pie chart

```python
import matplotlib.pyplot as plt

python_libraries = ['NumPy', 'Pandas', 'Matplotlib', 'Seaborn', 'Plotly', 'Scikit-learn']
skills = [9, 8, 7, 5, 3, 10]
colors = ['red','#ff9999','#66b3ff','#99ff99','#ffcc99', '#ffcc39']
plt.pie(skills, labels=python_libraries, colors=colors, explode=(0, 0, 0, 0, 0, 0.3), shadow=True, autopct='%1.2f%%')

plt.title('My Skills in Different Python Libraries')
plt.show()
```
